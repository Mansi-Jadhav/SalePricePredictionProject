```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();

# Through these plots, we can see that TV and Radio sales increase as we increase the amount spend.
# But newspaper sales are not really affected even though advertising amount is increased.
```

```python
X = df.drop('sales', axis=1)
```

```python
X.head()                  # Features
```

```python
y = df['sales']
```

```python
y.head()                 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# help(LinearRegression)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)                  # Train the model
```

```python
test_predictions = model.predict(X_test)         # Predict from model
test_predictions
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df['sales'].mean()
```

```python
sns.histplot(data=df, x='sales', bins=20)
```

```python
mean_absolute_error(y_test, test_predictions)       # Compare to mean of the df - 10% error approx
```

```python
mean_squared_error(y_test, test_predictions) 
```

```python
np.sqrt(mean_squared_error(y_test, test_predictions))   # Root mean squared error - Again around 10%
```
